#checking that the data is stationary or not
#ADF - AUGUMENTED DICKY FULLER TEST


```python
from statsmodels.tsa.stattools import adfuller
test_result = adfuller(df)
```

```python
#null hypo -- the data is not stationary

#alt hypo -- the data is stationary
```

```python
pval = test_result[1]
pval
```

```python
if pval <0.05:
  print('data is stationary')
else:
  print("data is not stationary")
```

As our data is not stationary we need to convert it into stationary

Logarithmic transformation

```python
first_log =np.log(df)
first_log = first_log.dropna()
first_log.plot()
```

```python
first_log
```

#rolling mean---mean_log


#rolling standard deviation--std_log

```python
mean_log = first_log.rolling(window = 12).mean()
std_log = first_log.rolling(window = 12).std()

plt.plot(first_log,color='blue',label ='orginal')
plt.plot(mean_log,color='red',label ='Rolling mean')
plt.plot(std_log,color='black',label ='Rolling std')
plt.legend(loc='best')
plt.title("rolling mean & rolling std")
```

```python
std_log.head(20)
```

```python
new_data = first_log - mean_log
new_data =  new_data.dropna()
new_data.head(20)
```

```python
plt.plot(new_data,color='blue',label ='orginal')
plt.plot(mean_log,color='red',label ='Rolling mean')
plt.plot(std_log,color='black',label ='Rolling std')
plt.legend(loc='best')
plt.title("rolling mean & rolling std")
```

```python
#ARIMA -AUTOREGRESSIVE INTEGRATED MOVING AVERAGE
#ARIMA-(p,d,q)
# p- order of lag
# d-degree  of differencing
# q-order of moving average

# p--how many past values you want to use for Prediction
# d---how many times you need to subtract past values to make the data steady
# q- how many past errors you want to use to improve  your predictions
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
from sklearn.metrics import *
```

```python
#itertools
import itertools
p = range(0,7)
d= range(1,2)
q=range(0,7)
pdq = list(itertools.product(p,d,q))
rmse = []
order1 = []
for i in pdq:
  model = ARIMA(train,order = i)
  model_fit = model.fit()
  pred = model_fit.predict(start = len(train),end = len(train)+len(test)-1) #start = 120, end = (120+13)-1 = 132 #0-132
  error = np.sqrt(mean_squared_error(test,pred))
  order1.append(i)
  rmse.append(error)
results = pd.DataFrame(index = order1,data = rmse,columns =['RMSE'])
```

```python
results.sort_values(by = 'RMSE',ascending = True)
```

```python
A_model = ARIMA(train,order=(6,1,6))
```

```python
new_data["ARIMA_Predictions"] = A_model.predict(start = len(train),end = len(train)+len(test)-1)
```

```python
new_data.tail(10)
```

```python
new_data[['#Passengers','ARIMA_Predictions']].plot()
```

```python
#SARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(train,order =(6,1,6),seasonal_order= (6,1,6,12))
model =model.fit()
```

```python
new_data['SARIMA_Predictions'] =model.predict(start = len(train),end = len(train)+len(test)-1)
```

```python
new_data[['#Passengers','SARIMA_Predictions']].plot()
```

```python
#predict for next 5 years---12*5 = 60months
future = model.forecast(steps = 60)
new_data.plot()
future.plot()
```
